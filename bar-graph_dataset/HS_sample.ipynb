```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

%matplotlib inline

plt.style.use('ggplot') 
```

```python
df = pd.read_csv("./source/HS_sample.csv", sep=",") # prepare sample list csv file 
df.head()
```
