```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

%matplotlib inline

plt.style.use('ggplot') 
```

```python
df = pd.read_csv("./source/HS_treatment.csv", sep=",")
df.head()
```

```python
df2 = df.set_index('Treatment time')
df2.head()
```

```python
sns.set_palette("bright")
 
plt.rcParams["figure.figsize"] = [30,10]   
df2.plot(kind = 'bar' )
plt.grid(which = "both", axis="x")
plt.xticks(fontsize=20, rotation=45)
plt.xlabel('Treatment time ', size = 30)
plt.ylabel('pair ', size = 30)
plt.legend(fontsize=40, ncol=1)
plt.tick_params(labelsize=20)
plt.ylim(0, 140)
plt.show
plt.savefig("./PNG_file/HS_treatment.png", format="png", dpi=200)
```
