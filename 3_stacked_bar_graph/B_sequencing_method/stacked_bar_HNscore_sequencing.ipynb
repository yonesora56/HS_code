```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

plt.style.use('ggplot') 
```

### sample type specific HN-score (Common up/downregulated genes in human and mouse )

```python
# HN-scores were calculated in human and mouse, respectively, and combined files were used

df = pd.read_csv("../source/common_up_down_regulated_gene_sequencing.csv", sep=",")
df.head()
```

```python
df2 = df.set_index('GeneName')
df2.head()
```

```python
# Specify colors manually as colors are cycled

color = [
    '#CD5C5C', # RNA-seq
    '#FFA07A', # TTchem-seq
    '#8B0000', # 4sU-seq
    '#FF1493', # TT-seq
    '#FF6347', # SLAM-seq
    '#FFD700', # PRO-seq
    '#FAFAD2', # MeRIP-seq
    '#EEE8AA', # m1A-seq
    '#D8BFD8', # m6Am-seq
    '#EE82EE', # mNET-seq
    '#FF00FF', # Ribo-seq
    '#556B2F', # RNA-seq(mouse)
    '#6B8E23', # miRNA-seq(mouse)
    '#20B2AA', # ncRNA-seq(mouse)
    '#00FFFF', # PRO-seq(mouse)
    '#48D1CC', # Timelapse-seq(mouse)
    '#1E90FF'  # RIbo-seq(mouse)
]


sns.set_palette(color)
 
plt.rcParams["figure.figsize"] = [140,80]   
df2.plot(kind = 'bar' , stacked=True)
x = plt.xlabel('GeneName', size = 80)
plt.xticks(fontsize=18, rotation=85)
y = plt.ylabel('HN-score', size = 90)
plt.legend(fontsize=150, ncol=3)
plt.tick_params(labelsize=60)
plt.ylim(-100, 400)
plt.savefig("../PNG_file/common_genes_sequencing_specific.png", format="png", dpi=100)
```
