```python
temp = pd.DataFrame()

# Extract unique column name 
sample_list = df1.columns.unique()

def process_columns(columns, df1, temp):
    for col in columns:
        thre = df1[col] >= 5
        temp[col + '-up'] = thre.sum(axis=1)
        thre = df1[col] <= 1/5
        temp[col + '-down'] = thre.sum(axis=1)
        thre = df1[col] >= 0
        temp[col + '-all'] = thre.sum(axis=1)

    for word in columns:
        temp["unchange_" + word] = temp[word + "-all"] - temp[word + "-up"] - temp[word + "-down"]  # unchange
        temp["HN5_" + word] = temp[word + "-up"] - temp[word + "-down"]  # HN-score (threshold 5)

# distinguish column type such as RNA-seq, ncRNA-seq
column_types = {'44℃:120min.': [], '43℃:60min.': [], '45℃:30min.': [], '42℃:60min.': [], '43℃:30min.': [], '39℃:1440min.': [], '42℃:30min.': [], '42.5℃:60min.': [], '45℃(+/-0.2):30min.': [], '41.5℃：420min.': [], '41.5℃:420min.': [], '42℃:240min.': [], '42℃:160min.': [], '42℃:45min.': [], '39℃:240min.': [], '42℃:60min.+recovery': [], '43℃:120min.': [], '43℃:180min.': [], '42℃:15min.': [], '43℃:105min.': [], '44℃:60min.': [], '42℃:65min.': [], '43℃:240min.': [], '42℃:9×60min.': []}
for col in sample_list:
    for key in column_types.keys():
        if key in col and col not in column_types[key]: # Run only if the column name contains a specific key and the column has not already been added to the list
            column_types[key].append(col)

# Perform processing for each column type
for key, columns in column_types.items():
    process_columns(columns, df1, temp)

all_columns = temp.filter(regex='-all$')  #ALL pair (for check)
temp["all(check)"] = all_columns.sum(axis=1)

total_columns = temp.filter(regex='^HN5')  #Total_HN-score
temp["Total_HN-score"] = total_columns.sum(axis=1)

temp

```

```python
tempHN5= temp.sort_values(by= "Total_HN-score", ascending=False)
tempHN5
```
