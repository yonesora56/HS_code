```python
df1.rename(mapper=condition, axis=1, inplace= True)
df1
```

```python
sample_list = df1.columns.unique()
column_types = {col: [] for col in sample_list}
print(column_types)
```

```python
temp = pd.DataFrame()

# Extract unique column name 
sample_list = df1.columns.unique()

for col in sample_list:  # distinct UP, DOWN, ALL from df1
    thre = df1.filter(like = col)>=5 
    temp[col + '-up' ] = thre.sum(axis=1)
    thre = df1.filter(like = col)<=1/5
    temp[col + '-down' ] = thre.sum(axis=1)
    thre = df1.filter(like = col) >= 0
    temp[col  + '-all'] = thre.sum(axis=1)

all_columns = temp.filter(regex='-all$')  #ALL pair (for check)
temp["all(check)"] = all_columns.sum(axis=1)

thre =df1>= 0  # ALL pair
temp["all"] = thre.sum(axis=1)

for word in sample_list:
    temp["unchange_" + word] = temp[word + "-all"] - temp[word + "-up"] - temp[word + "-down"] # unchnage
    temp["HN5_" + word] =  temp[word + "-up"] - temp[word + "-down"] #HN-score (threshold 5)


total_columns = temp.filter(regex='^HN5')  #Total_HN-score
temp["Total_HN-score"] = total_columns.sum(axis=1)

temp
```

```python
tempHN5= temp.sort_values(by= "Total_HN-score", ascending=False)
tempHN5
```
