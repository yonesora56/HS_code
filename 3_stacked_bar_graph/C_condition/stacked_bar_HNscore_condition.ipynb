```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

plt.style.use('ggplot') 
```

### sample type specific HN-score (Common up/downregulated genes in human and mouse )

```python
# HN-scores were calculated in human and mouse, respectively, and combined files were used

df = pd.read_csv("../source/common_up_down_regulated_gene_condition.csv", sep=",")
df.head()
```

```python
df2 = df.set_index('GeneName')
df2.head()
```

```python
# Specify colors manually as colors are cycled

color = ['#CD5C5C', '#FA8072', '#FFA07A', '#FF0000', '#8B0000', '#FFB6C1', '#FF1493', '#DB7093'
       , '#FF6347', '#FF8C00', '#FFD700', '#FFFFE0', '#FAFAD2', '#FFE4B5', '#EEE8AA', '#BDB76B'
        , '#D8BFD8', '#EE82EE', '#FF00FF', '#BA55D3', '#9966CC', '#9400D3', '#8B008B', '#4B0082'
        , '#483D8B', '#32CD32', '#00FF00', '#98FB98', '#00FA9A', '#3CB371', '#228B22' , '#006400'
        , '#6B8E23', '#556B2F', '#20B2AA' , '#00FFFF', '#48D1CC', '#B0E0E6' , '#1E90FF' , '#0000CD'
        , '#D2691E' , '#800000' , '#F8B400']


sns.set_palette(color)
 
plt.rcParams["figure.figsize"] = [140,80]   
df2.plot(kind = 'bar' , stacked=True)
x = plt.xlabel('GeneName', size = 80)
plt.xticks(fontsize=18, rotation=85)
y = plt.ylabel('HN-score', size = 90)
plt.legend(fontsize=80, ncol=3)
plt.tick_params(labelsize=60)
plt.ylim(-100, 400)
plt.savefig("../PNG_file/common_genes_condition_specific.png", format="png", dpi=100)
```
