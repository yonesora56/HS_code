```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
import pandas as pd
import numpy as np

rcParams["font.family"]     = "Arial"
rcParams["font.sans-serif"] = "Hiragino Maru Gothic Pro"
plt.style.use("ggplot")
```

```python
# HN-score csv file created using HN-score script (https://github.com/no85j/hypoxia_code)
HNscore = pd.read_csv("./source/HN-score_human_HN5_all.csv") #this example using human HN-score csv file
HNscore
```

# All gene

```python
# Extract the columns of HN-score(HN5) calculated with expression ratio 5x, 1/5x

HN5 = HNscore.iloc[: , 17]
HN5  # Output for confirmation
```

```python
listx=list(range(1, 19706))
All = {"X":     [listx],  
        "Y":             [HN5]
      }
```

## Upregulated gene

```python
UP5 = np.array(range(1, 501)) 
UP5
```

```python
HN5up = HNscore.iloc[0:500 , 17]
HN5up
```

```python
upregulated = {"X": [UP5], 
     "Y" : [HN5up]
        }
```

## Downregulated gene

```python
DOWN5 = np.array(range(19205 ,19705))  
DOWN5
```

```python
HN5down = HNscore.iloc[19205:19705 , 17]
HN5down
```

```python
downregulated = {"X": [DOWN5], 
        "Y": [HN5down]
        }
```

```python
ax = plt.axes()
ax.set_facecolor('white')

#All genes
plt.scatter(
            All["X"],     
            All["Y"],    
            marker = "o",        
            s = 30,      
            c = "grey",       
            edgecolor = "black",   
            linewidth = 0.01,    
            alpha = 0.5,        
            label = "all" 
            )
# upregulated genes
plt.scatter(
            upregulated["X"],          
            upregulated["Y"],     
            s = 100,          
            c = "red",       
            edgecolor = "black",    
            linewidth = 1,      
            alpha = 1,         
            label = "upregulated"   
            )

#downregulated genes
plt.scatter(
            downregulated["X"],        
            downregulated["Y"],    
            s = 100,          
            c = "blue",      
            edgecolor = "white",   
            linewidth = 1,      
            alpha = 1,            
            label = "downregulated"   
            )

plt.grid(color='b', linestyle=':', linewidth=0.3)
plt.rcParams["figure.figsize"] = [7, 13]  
plt.xlabel("Gene ranking based on HN-score", fontsize=20) 
plt.ylabel("HN-score (human) ", fontsize=20)
plt.tick_params(labelsize=14)
plt.savefig("./PNG_file/scatter-plot_HN5_human.png", format="png", dpi=500)
```
