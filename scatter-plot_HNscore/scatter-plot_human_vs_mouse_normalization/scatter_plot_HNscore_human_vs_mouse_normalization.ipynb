```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
import pandas as pd
import numpy as np
import datetime

rcParams["font.family"]     = "Arial"
rcParams["font.sans-serif"] = "Hiragino Maru Gothic Pro"
plt.style.use("ggplot")
```

## Normalization (0~1)

```python
data = pd.read_csv("../HN-score_HN5_convert.csv")
data1 = pd.DataFrame(data)
data1
```

```python
# human and mouse HN-score extraction
data2 = data1.iloc[:, 2:4]
data2
```

```python
# 0 to 1 normalization

from sklearn import preprocessing

mm = preprocessing.MinMaxScaler()

data3 = pd.DataFrame(mm.fit_transform(data2),
                          index=data2.index, columns=data2.columns)
data3
```

```python
data3.insert(0, "human_symbol" , data1["human symbol"], )
data3
```

```python
data3.to_csv('./HN-score_human_mouse_HN5_normalization.csv', sep=",")
```

```python
UP = pd.read_csv("./HN-score_human_mouse_HN5_normalization_UP.csv")
UP
```

```python
DOWN = pd.read_csv("./HN-score_human_mouse_HN5_normalization_DOWN.csv")
DOWN
```

```python
ax = plt.axes()
ax.set_facecolor('white')
plt.scatter(
            data3["human"],            
            data3["mouse"],          
            marker = "o",        
            s = 15,              
            c = "grey",      
            edgecolor = "black",   
            linewidth = 0.01,    
            alpha = 0.5,        
            label = "all" 
            )

plt.scatter(
            UP["human"],           
            UP["mouse"],      
            s = 40,         
            c = "red",         
            edgecolor = "white",   
            linewidth = 1,    
            alpha = 1,       
            label = "upregulate"   
            )

plt.scatter(
            DOWN["human"],            
            DOWN["mouse"],     
            s = 40,            
            c = "blue",     
            edgecolor = "white",   
            linewidth = 1,      
            alpha = 1,      
            label = "upregulate"   
            )

plt.show
plt.grid(color='b', linestyle=':', linewidth=0.1)
plt.grid(which = "both", axis="y")
plt.rcParams["figure.figsize"] = [15, 11]  
plt.xlabel("HN-score (human) normalization ", fontsize=15) 
plt.ylabel(" HN-score (mouse) normalization", fontsize=15)
plt.tick_params(labelsize=9)
plt.savefig("./HNscore_normalization.png", format="png", dpi=500)
```
