```python
ax = plt.axes()
ax.set_facecolor('white')
plt.scatter(
            All["human"],           
            All["mouse"],           
            marker = "o",        
            s = 15,                
            c = "grey",           
            edgecolor = "black", 
            linewidth = 0.01,        
            alpha = 0.5,           
            label = "all"  
            )

plt.scatter(
            upregulated["X"],           
            upregulated["Y"],         
            s = 40,              
            c = "red",           
            edgecolor = "white", 
            linewidth = 1,    
            alpha = 1,        
            label = "upregulate"  
            )

plt.scatter(
            downregulated["X"],   
            downregulated["Y"],  
            s = 40,            
            c = "blue",      
            edgecolor = "white",   
            linewidth = 1,      
            alpha = 1,       
            label = "crowded"
            )

plt.grid(color='b', linestyle=':', linewidth=0.1)
ax.set_xlim(-50, 250)
ax.set_ylim(-50, 200)
ax.minorticks_on()
plt.grid(which = "both", axis="y")
plt.rcParams["figure.figsize"] = [15, 11]  
plt.xlabel("HN-score (human) ", fontsize=15) 
plt.ylabel(" HN-score (mouse) ", fontsize=15)
plt.tick_params(labelsize=9)
plt.savefig("./PNG_file/scatter-plot_HN5_human_vs_mouse.png", format="png", dpi=500)
```
